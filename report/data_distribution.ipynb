```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    'Poet': [
        'Abbasin Yousuf', 'Ajmal Khattak', 'Allama Abdul Hai', 'Aziz Mazerwal', 'Ghani Khan', 'Hamza Baba',
        'Javed Ahmedzai', 'Karan Khan', 'Khaliq Ziari', 'Khatir Afridi', 'Khushal Khan Khattak', 'Matiullah Turab',
        'Mumtaz Orakazi', 'Munir Jan', 'Naeem Ahmed', 'Rabia Mumtaz', 'Rahman Baba', 'Rehmat Shah', 'Sahib Shah Sabir',
        'Shabbir Khan Durrani', 'Shakir Orakzai', 'Salim Riaz', 'Shoaib Khan Khattak'
    ],
    'Couplet Count': [
        1875, 780, 2380, 30, 1399, 1177, 561, 979, 40, 1436, 4257, 1192, 1645, 1169, 856, 572, 2202, 1534, 1051,
        1056, 569, 232, 615
    ]
}
```

```python
# Creating a DataFrame
df = pd.DataFrame(data)

# Setting the aesthetic style of the plots
sns.set_style("whitegrid")
```

```python
# Plotting a histogram
plt.figure(figsize=(12, 6))
sns.histplot(df['Couplet Count'], kde=True)
plt.title('Distribution of Couplet Counts')
plt.xlabel('Couplet Count')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plotting a barplot
plt.figure(figsize=(14, 8))
# sns.barplot(x='Couplet Count', y='Poet', data=df, palette='viridis', hue='Poet')
sns.barplot(x='Couplet Count', y='Poet', data=df, palette='viridis', hue='Couplet Count')
plt.title('Number of Couplets by Poet')
plt.xlabel('Couplet Count')
plt.ylabel('Poet')
plt.show()
```

```python
# Plotting a scatter plot
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Poet', y='Couplet Count', data=df, hue='Couplet Count', palette='viridis', s=100, legend=False)
plt.xticks(rotation=90)
plt.title('Scatter Plot of Couplets by Poet')
plt.xlabel('Poet')
plt.ylabel('Couplet Count')
plt.show()
```
